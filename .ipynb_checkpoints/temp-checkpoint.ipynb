```python
import menzalib as mz
import numpy as np
import pylab as plt
import scipy.optimize
import scipy.integrate
import scipy.special
import numdifftools.nd_algopy as nda
from sklearn import linear_model as lin
from inspect import signature
```

```python
def f(t, a):
    return a*np.array([np.cos(t)*np.sin(t)**2, np.sin(2*t)])

def L(t, df, a):
    df = df(t, a)
    return np.sqrt(df[0]**2+df[1]**2)

a=1
df = nda.Derivative(f)
t = np.linspace(0,np.pi/2, 1000)
x,y = f(t, a)
print(np.max(np.sqrt(x**2 + y**2)))
print(scipy.integrate.quad(L, 0, np.pi/2, args=(df, a)))

for i in range(10):
    a = i/10
    x,y = f(t, a)
    plt.plot(x,y)
plt.show()
```

```python
N = 50000

x = np.random.uniform(0,1,(N, 2, 1))
dist = np.linalg.norm(x[:, 0]-x[:, 1], axis=1)

# plt.hist(dist, bins=50)
# plt.show()
print(np.mean(dist)-1/3, np.std(dist)/np.sqrt(N))
```

```python
%matplotlib qt

N = 400

def f(x, a,b,c,d):
    return a + b*x + c*x**2 + d*x**3

def CFlevel(X, pcov):
    df = np.zeros(N)
    
    for i in range(N):
        x = np.array([1,X[i],X[i]**2,X[i]**3])
        df[i] = 2*np.sqrt(np.linalg.multi_dot((x, pcov, x.T)))
    return df

x = np.linspace(-2,2, N)
y = f(x, 1.396, 1.941, -0.1216, 0.1535) + np.random.normal(0, 1, N)

popt, pcov = scipy.optimize.curve_fit(f, x, y)
print(popt, np.sqrt(np.diag(pcov)))
df = CFlevel(x, pcov)

plt.plot(x,y, '.')
plt.plot(x, f(x, *popt))
plt.plot(x, f(x, *popt)+df)
plt.plot(x, f(x, *popt)-df)
plt.show()
```
